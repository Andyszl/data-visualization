```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
```

### 基本图

```python
fig = plt.figure(figsize=(8,6))
plt.boxplot(data,notch=False,sym='s',vert=True)

plt.xticks([y+1 for y in range(len(data))],['x1','x2','x3'])
plt.xlabel('x')
plt.title('box plot')
           
```

### 颜色设置

```python
fig = plt.figure(figsize=(8,6))
bplot=plt.boxplot(data,notch=False,sym='s',vert=True)

plt.xticks([y+1 for y in range(len(data))],['x1','x2','x3'])
plt.xlabel('x')
plt.title('box plot')

for components in bplot.keys():
    for line in bplot[components]:
        line.set_color('blue')
```

### 横置

```python
fig = plt.figure(figsize = (8,6))
plt.boxplot(data,notch=False,sym='s',vert=False)

plt.yticks([y+1 for y in range(len(data))],['x1','x2','x3'])
plt.ylabel('x')
plt.title('box plot')
```

### notch=True

```python
fig = plt.figure(figsize = (8,6))
plt.boxplot(data,notch=True,sym='s',vert=False)

plt.yticks([y+1 for y in range(len(data))],['x1','x2','x3'])
plt.ylabel('x')
plt.title('box plot')
```

### 分别配色

```python
fig = plt.figure(figsize = (8,6))
bplot = plt.boxplot(data,notch=False,sym='s',vert=True,patch_artist=True)

plt.xticks([y+1 for y in range(len(data))],['x1','x2','x3'])
plt.xlabel('x')
plt.title('box plot')

colors = ['pink','lightblue','lightgreen']
for pathch,color in zip(bplot['boxes'],colors):
    pathch.set_facecolor(color)
```

### 小提琴图violinplot

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,5))
axes[0].violinplot(data,showmeans=False,showmedians=True)
axes[0].set_title('violin plot')

axes[1].boxplot(data)
axes[1].set_title('box plot')

```

##### 加上网格线及坐标轴标签

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,5))
axes[0].violinplot(data,showmeans=False,showmedians=True)
axes[0].set_title('violin plot')

axes[1].boxplot(data)
axes[1].set_title('box plot')
for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y+1 for y in range(len(data))])
plt.setp(axes,xticks=[y+1 for y in range(len(data))],xticklabels=['x1','x2','x3'])
```

##### 坐标轴线显示

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,5))
axes[0].violinplot(data,showmeans=False,showmedians=True)
axes[0].set_title('violin plot')

axes[1].boxplot(data)
axes[1].set_title('box plot')
for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y+1 for y in range(len(data))])
plt.setp(axes,xticks=[y+1 for y in range(len(data))],xticklabels=['x1','x2','x3'])
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)
```
