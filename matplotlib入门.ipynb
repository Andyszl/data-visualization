### 线条类型

```python
#构造数据
a_numpy = np.arange(0,2,0.05)
```

```python
plt.figure(figsize=(10,6))#设置画布大小
plt.plot(a_numpy,a_numpy,'r--')
plt.plot(a_numpy,a_numpy**2,'bs')
plt.plot(a_numpy,a_numpy**3,'go')
```

### 子图

```python
#构造数据
x = np.linspace(-10,10)
y = np.sin(x)
```

```python
#三行两列数据
#第一个图
plt.subplot(321)
plt.plot(x,y,color='r')
#第四个图
plt.subplot(324)
plt.plot(x,y,color='b')
#第五个图
plt.subplot(325)
plt.plot(x,y,color='g')
```

### 线条基本设置

```python
plt.figure(figsize=(10,8))
#三行两列数据
#第一个图，原始图形
plt.subplot(321)
plt.plot(x,y,color='g')

#第二个图,线条格式
plt.subplot(322)
plt.plot(x,y,color='r',linestyle=':')

#第三个图,数据点格式
plt.subplot(323)
plt.plot(x,y,color='g',marker = 'o',markerfacecolor='r',markersize = 10)

#第四个图，透明度
plt.subplot(324)
plt.plot(x,y,color='g', alpha = 0.4)

#第五个图,线条粗细
plt.subplot(325)
plt.plot(x,y,color='r',linewidth = 3.0)

#第六个图,点图
plt.subplot(326)
plt.plot(x,y,'ro')
```

### 注释

```python
plt.plot(x,y,color='b',linestyle=':',marker = 'o',markerfacecolor='r',markersize = 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('andy')
plt.text(0,0,'andy')#指定位置加注释
plt.grid(True)#加上轴线方格
plt.annotate('andy',xy=(-5,0),xytext=(-2,-0.5),arrowprops = dict(facecolor='black',shrink=0.05,headlength= 20,headwidth = 20))#shrink放缩，headwidth宽度，headlength长度
```

```python
import math   
#极坐标图
t=np.linspace(0,2*math.pi) 
x=3*np.sin(t); 
y=4*np.cos(t) 
plt.plot(x,y); 
plt.text(-1,0,r'$\frac{x^2}{3^2}+\frac{y^2}{4^2}=1$',fontsize=20)
```
